```python
import matplotlib.pyplot as plt

# Use dark background
plt.style.use('dark_background')

# Data
years = list(range(1, 23))  # Years 1 to 22
execution_times = [114, 123, 165, 216, 259, 280, 298, 347, 399, 455, 493, 
538, 577, 621, 676, 721, 774, 827, 869, 994, 1140, 1259]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years, execution_times, marker='o', color='cyan')

# Annotate points slightly above with bigger font
for x, y in zip(years, execution_times):
    plt.text(x, y + 30, str(y), ha='center', fontsize=11, color='white')

# Labels and title
plt.title('Execution Time: 1 to 22 Years Data', color='white')
plt.xlabel('Years', color='white')
plt.ylabel('Execution Time (Seconds)', color='white')

# X-axis ticks
plt.xticks(ticks=years, labels=[str(year) for year in years], color='white')
plt.yticks(color='white')

# Grid with transparency
plt.grid(True, color='gray', alpha=0.3)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Use dark background
plt.style.use('dark_background')

# Data
years = list(range(1, 23))  # Years 1 to 22
execution_times_seconds = [114, 123, 165, 216, 259, 280, 298, 347, 399, 455, 493, 
                           538, 577, 621, 676, 721, 774, 827, 869, 994, 1140, 1259]

# Convert to minutes (rounded to 2 decimal places)
execution_times_minutes = [round(sec / 60, 2) for sec in execution_times_seconds]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years, execution_times_minutes, marker='o', color='cyan')

# Annotate points slightly above with bigger font
for x, y in zip(years, execution_times_minutes):
    plt.text(x, y + 0.5, str(y), ha='center', fontsize=11, color='white')

# Labels and title
plt.title('Execution Time: 1 to 22 Years Data', color='white')
plt.xlabel('Years', color='white')
plt.ylabel('Execution Time (Minutes)', color='white')

# X-axis ticks
plt.xticks(ticks=years, labels=[str(year) for year in years], color='white')
plt.yticks(color='white')

# Grid with transparency
plt.grid(True, color='gray', alpha=0.3)

# Show plot
plt.tight_layout()
plt.show()
```
